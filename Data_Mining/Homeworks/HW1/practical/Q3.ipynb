### Question4
Display the two characteristics of longitude and latitude on the map using location data and specify the density of the areas where there are more houses or on the map.

```python
fig = plt.figure(figsize=(12,9))

m = Basemap(projection = 'mill', llcrnrlat = -90, urcrnrlat = 90, llcrnrlon = -180, urcrnrlon = 180, resolution = 'c')
# drawing the coastline
m.drawcoastlines()
m.drawcountries(color='gray')
m.drawstates(color='gray')

lat = houses['latitude'].tolist()
lon = houses['longitude'].tolist()
# plotting the map
m.scatter(lon, lat, latlon = True, s = 10, c = 'red', marker = 'o', alpha = 1)

plt.show()
```

```python
# plotting map using plotly
plt.figure(figsize=(15,9))
fig = px.scatter_geo(houses, lat = 'latitude', lon = 'longitude')

# setting title for the map
fig.update_layout(title = 'Houses', title_x = 0.5)
```
