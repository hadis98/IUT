### Question5
Add a new column called clock_rate to the dataset and for phones whose clock_speed is equal to or less than 1, set the value of this column to Low, for those that are greater than 1 and less than 2, set the value of this column to Mid, and for those that clock_speed is more than 2. Set the new column value to High

```python
def set_clock_speed(clock_speed):
    if clock_speed <=1 :
        return "Low"
    elif clock_speed >1 and clock_speed <2:
        return "Mid"
    else: 
        return "High"
```

```python
smartphones["clock_rate"] = smartphones["clock_speed"].apply(set_clock_speed)
```
