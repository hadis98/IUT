## Question10
What are other strategies for replacing missing values? Display the output of each method.

1.Drop missing values: This strategy involves removing any rows or columns that contain missing values. This can be done using the dropna() method of a pandas DataFrame.

2.Mean/Median/Mode imputation: In this strategy, missing values are replaced with the mean, median, or mode value of the respective feature/column. This can be done using the fillna() method of a pandas DataFrame.

3.Interpolation: In this strategy, missing values are replaced with interpolated values based on the values of other data points. This can be done using the interpolate() method of a pandas DataFrame.

4.Machine learning algorithms: Missing values can also be filled using machine learning algorithms such as K-Nearest Neighbors, Decision Trees, Random Forests, etc.

```python
#3.Interpolation

new_df4 = df.iloc[:,0:5]
print('\nbefore using interpolate\n')
print(new_df4.isnull().sum()) # check number of missing values
new_df4.interpolate(method='linear',inplace=True)
print('\nafter using interpolate\n')
print(new_df4.isnull().sum()) # confirm missing values have been filled with interpolated values
```

```python
#4.Machine learning algorithms:
from sklearn.impute import KNNImputer

new_df5 = df.iloc[:,0:5]
print('\nbefore using KNNImputer\n')
print(new_df5.isnull().sum()) # check number of missing values
imputer = KNNImputer()
new_df5_imputed = imputer.fit_transform(new_df5)
# new_df5_imputed.columns = new_df5.columns
print('\nafter using KNNImputer\n')
print(pd.DataFrame(new_df5_imputed).isnull().sum()) # confirm there are no more missing values
```

## Question11
Which algorithms are resistant to missing values? Explain.

### answer
Some machine learning algorithms are resistant to missing values, meaning that they can handle datasets with missing values without requiring imputation or removal of the missing data. Here are a few examples:

Tree-based models (Decision Trees, Random Forests, Gradient Boosted Trees) - These models partition the feature space into smaller subsets based on the feature value that provides the highest information gain at each split. Since these models work on individual features at each split, they can handle missing values by simply treating them as a separate category or by making splits based on non-missing data.

Support Vector Machines (SVMs) - SVMs work by finding the hyperplane that maximally separates the different classes in the feature space. When missing values are present, SVMs can ignore those features when constructing the boundary, as long as other informative features are available.

K-Nearest Neighbors (KNN) - KNN is a non-parametric algorithm that classifies a new observation based on the class of its K nearest neighbors in the training set. Since KNN operates on distances between observations and doesn't require explicit modeling like other methods, it can handle missing values by simply ignoring them during distance calculations.

Naive Bayes - Naive Bayes is a probabilistic classifier that assumes independence between features given the class. When missing values are present, Naive Bayes can simply ignore them during calculations of conditional probabilities.

It's important to note, however, that if the percentage of missing values is too high, even these algorithms may not perform well. In general, it's good practice to try to minimize the amount of missing data as much as possible before applying any machine learning algorithm.
